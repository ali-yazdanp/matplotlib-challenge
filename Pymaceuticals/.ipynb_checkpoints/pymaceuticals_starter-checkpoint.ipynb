```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

combine_pd = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
combine_pd
```

```python
# Checking the number of mice.
no_mice = combine_pd["Mouse ID"].nunique()
no_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_mice_by_ID = combine_pd.loc[combine_pd.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate_mice_by_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

duplicate_mice_by_ID_df = combine_pd.loc[combine_pd["Mouse ID"] == "g989", :]
duplicate_mice_by_ID_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_of_duplicates_df = combine_pd[combine_pd['Mouse ID'].isin(duplicate_mice_by_ID)==False]
cleaned_of_duplicates_df
```

```python
# Checking the number of mice in the clean DataFrame.
no_mice_clean = cleaned_of_duplicates_df["Mouse ID"].nunique()
no_mice_clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumour volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumour volume. 

Mean = cleaned_of_duplicates_df['Tumour Volume (mm3)'].groupby(cleaned_of_duplicates_df['Drug Regimen']).mean()
Median = cleaned_of_duplicates_df['Tumour Volume (mm3)'].groupby(cleaned_of_duplicates_df['Drug Regimen']).median()
Variance = cleaned_of_duplicates_df['Tumour Volume (mm3)'].groupby(cleaned_of_duplicates_df['Drug Regimen']).var()
Std_Deviation = cleaned_of_duplicates_df['Tumour Volume (mm3)'].groupby(cleaned_of_duplicates_df['Drug Regimen']).std()
SEM = cleaned_of_duplicates_df['Tumour Volume (mm3)'].groupby(cleaned_of_duplicates_df['Drug Regimen']).sem()

# Assemble the resulting series into a single summary dataframe.

Summary_statistics = pd.DataFrame({"Tumour Volume (mm3) Mean":Mean, 
                            "Tumour Volume (mm3) Median":Median, 
                           "Tumour Volume (mm3) Variance":Variance, 
                           "Tumour Volume (mm3) Standard Deviation.":Std_Deviation, 
                           "Tumour Volume (mm3) SEM.":SEM})
Summary_statistics
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumour volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

Summary_statistics_aggregation =  cleaned_of_duplicates_df.groupby(['Drug Regimen'])[['Tumour Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
Summary_statistics_aggregation
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

Capomulin_df = cleaned_of_duplicates_df.loc[cleaned_of_duplicates_df["Drug Regimen"] == "Capomulin",:]
Ramicane_df = cleaned_of_duplicates_df.loc[cleaned_of_duplicates_df["Drug Regimen"] == "Ramicane", :]
Infubinol_df = cleaned_of_duplicates_df.loc[cleaned_of_duplicates_df["Drug Regimen"] == "Infubinol", :]
Ceftamin_df = cleaned_of_duplicates_df.loc[cleaned_of_duplicates_df["Drug Regimen"] == "Ceftamin", :]
```

```python
# Start by getting the last (greatest) timepoint for each mouse

Capomulin_last = Capomulin_df.groupby('Mouse ID').max()['Timepoint']
Capomulin_TV = pd.DataFrame(Capomulin_last)
Capomulin_merge = pd.merge(Capomulin_TV, cleaned_of_duplicates_df, on=("Mouse ID","Timepoint"),how="left")


Ramicane_last = Ramicane_df.groupby('Mouse ID').max()['Timepoint']
Ramicane_TV = pd.DataFrame(Ramicane_last)
Ramicane_merge = pd.merge(Ramicane_TV, cleaned_of_duplicates_df, on=("Mouse ID","Timepoint"),how="left")

Infubinol_last = Infubinol_df.groupby('Mouse ID').max()['Timepoint']
Infubinol_TV = pd.DataFrame(Infubinol_last)
Infubinol_merge = pd.merge(Infubinol_TV, cleaned_of_duplicates_df, on=("Mouse ID","Timepoint"),how="left")

Ceftamin_last = Ceftamin_df.groupby('Mouse ID').max()['Timepoint']
Ceftamin_TV = pd.DataFrame(Ceftamin_last)
Ceftamin_merge = pd.merge(Ceftamin_TV, cleaned_of_duplicates_df, on=("Mouse ID","Timepoint"),how="left")
```

```python
# Merging results for max time point for all drugs 

Drug_Regimen = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
Drug_Regimen_TV = combine_pd[combine_pd["Drug Regimen"].isin(Drug_Regimen)]

Max_timepoint = Drug_Regimen_TV.groupby(["Drug Regimen", "Mouse ID"]).agg(tumor_size=("Tumour Volume (mm3)", lambda x: x.iloc[-1]))
Max_timepoint = Max_timepoint.stack(level=0).unstack(level=0)
Max_timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
for Drug_Regimen_TV in Drug_Regimen:
    print(Drug_Regimen_TV)
```

```python
treatment = 0
for Drug_Regimen_TV in Drug_Regimen:
    quartiles = Max_timepoint[Drug_Regimen_TV].quantile([.25,.5,.75]).round(2)
    lowerq = quartiles[0.25].round(2)
    upperq = quartiles[0.75].round(2)
    iqr = round(upperq-lowerq,2)
    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)

    if treatment == 0:
        print(f"------------------------------------------------------------")
    print(f"The lower quartile of {Drug_Regimen_TV} treatments is: {lowerq}")
    print(f"The upper quartile of {Drug_Regimen_TV} treatments is: {upperq}")
    print(f"The interquartile range of {Drug_Regimen_TV} treatments is: {iqr}")
    print(f"Values below {lower_bound} could be {Drug_Regimen_TV} outliers.")
    print(f"Values above {upper_bound} could be {Drug_Regimen_TV} outliers.")
    print(f"------------------------------------------------------------")
    treatment+=1
```

```python
# Generate a box plot of the final tumour volume of each mouse across four regimens of interest
box_plot_list = []
for Drug_Regimen_TV in Drug_Regimen:
    box_plot_list.append(list(Max_timepoint[Drug_Regimen_TV].dropna()))
    
fig = plt.figure()
plt.xlabel("Drug Regimen")
plt.xticks([1,2,3,4], Drug_Regimen, rotation=45)
plt.ylabel("Tumour Volume (mm3)")
plt.title("Tumour Volume (mm3) of each mouse across the four Drug Regimen")
plt.boxplot(box_plot_list)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient 

# for mouse weight and average tumour volume for the Capomulin regimen

Correl = st.pearsonr(Capomulin_avg_TV['Weight (g)'],Capomulin_avg_TV['Tumour Volume (mm3)'])
print(f"The correlation between mouse weight and the average tumor volume for the Capomulin regimen is {round(Correl[0],3)}")
```

```python
# Calculate the  linear regression model 

(slope, intercept,rvalue, pvalue, stderr)= st.linregress(Capomulin_avg_TV["Weight (g)"],Capomulin_avg_TV["Tumour Volume (mm3)"])
regress=Capomulin_avg_TV["Weight (g)"]* slope + intercept

plt.scatter(Capomulin_avg_TV['Weight (g)'],Capomulin_avg_TV['Tumour Volume (mm3)'], color="red")
plt.title('Mouse Weight (g) Vs. Average Tumour Volume (mm3)')
plt.xlabel('Weight (g)')
plt.ylabel('Averag Tumor Volume (mm3)')
plt.plot(Capomulin_avg_TV["Weight (g)"], regress, color='blue')
print(f"The r^2 is: {round(rvalue**2,3)}")
plt.show()
```
